```python
import matplotlib.pyplot as plt
```

```python
size = 70692
init_size = [1.0, 0.9995, 0.999, 0.995, 0.993, 0.99]
n_updates = [int((1-t)*size) for t in init_size]
n_updates[0] = 1
no_rounding = [1, 2, 4, 8, 15, 36]
DD_rounding_1_digit = [1248225, 1404483, 1571437, 2555418, 2884541, 3628614]

DD_2_digit = [1790988, 3608463, 9949622, 15963689, 32309354, 65473071]

DD_no_rounding = [1604638, 11970964, 29084940, 70626309, 139800523, 346604638, ]

sklearn = [18739, 862853, 1580774, 5228580, 6958040, 9626720]
data = [DD_rounding_1_digit, DD_2_digit, sklearn]
labels = ["DD: rounding 2 Digits", "DD: rounding 1 Digits","Sklearn",]
```

```python
def plot(x, y, label, pos):
    markers = ['o', 's', 'D', '^', 'v']
    colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'orange', 'purple', 'brown', 'pink']
    plt.plot(x, y, marker=markers[0], linestyle='-', color=colors[pos], label=label)
```

```python
plt.figure(figsize=(12,5))


plot(no_rounding, DD_no_rounding, "DD: no rounding", -1)
for i, measures in enumerate(data):
    n = len(measures)
    plot(n_updates[:n], measures, labels[i], i)

plt.xscale('log')  # Log scale for better visualization
plt.yscale('log')  # Log scale since values vary a lot
plt.xlabel('Number of updates')
plt.ylabel('Time in qs')
plt.title('Performance Comparison of DD vs sklearn for Diabetes Dataset Pipeline')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```
